```python
import numpy as np
import pandas as pd
import os
import re
import plotly.express as px
import csv
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
import statsmodels.api as sm
warnings.filterwarnings("ignore")
```

```python
df_prec_month = pd.read_csv("df_prec_Dec.csv")
```

```python
df_prec_month = df_prec_month.loc[:,["Year","Area","Value"]]
df_prec_month
```

```python
year_list = df_prec_month.Year.unique()

weights = [0.0013, 0.0039, 0.0093, 0.0020, 0.0031, 0.1224, 0.0089, 0.0029, 0.0029, 0.0015, 0.7662, 0.0064, 0.0219, 0.0271, 0.0201]
USSR = ['Republic of Moldova', 'Belarus','Estonia', 'Latvia', 'Lithuania', 'Kazakhstan', 'Kyrgyzstan', 'Tajikistan', 'Turkmenistan', 'Uzbekistan', 'Russian Federation', 'Armenia', 'Azerbaijan', 'Georgia', 'Ukraine']

for y in year_list:
  Value = 0
  for i in list(range(0,len(USSR))):
    country = USSR[i]
    Value += weights[i]*df_prec_month.loc[(df_prec_month.Area == country)&(df_prec_month.Year == y),"Value"].sum()
  new_row = {"Year": y, "Area": "USSR", "Value": Value}
  df_prec_month = df_prec_month.append(new_row,ignore_index=True)

df_prec_month = df_prec_month.loc[~df_prec_month.Area.isin(USSR)]
```
